```python
import os
import random
import shutil

# Source folders
source_folder_cat = r"C:\aa projects\machine learning\Image data -Deep Learning\dog vs cat\kagglecatsanddogs_3367a\PetImages\resized_cat"
source_folder_dog = r"C:\aa projects\machine learning\Image data -Deep Learning\dog vs cat\kagglecatsanddogs_3367a\PetImages\resized_dog"

# Destination folder
destination_folder = r"C:\aa projects\machine learning\Image data -Deep Learning\dog vs cat\kagglecatsanddogs_3367a\PetImages\train_data_cat_dog"

# Copy 1000 random cat images
cat_images = random.sample(os.listdir(source_folder_cat), 1000)
for cat_image in cat_images:
    src_path = os.path.join(source_folder_cat, cat_image)
    dst_path = os.path.join(destination_folder, f"cat_{cat_image}")
    shutil.copy(src_path, dst_path)

# Copy 1000 random dog images
dog_images = random.sample(os.listdir(source_folder_dog), 1000)
for dog_image in dog_images:
    src_path = os.path.join(source_folder_dog, dog_image)
    dst_path = os.path.join(destination_folder, f"dog_{dog_image}")
    shutil.copy(src_path, dst_path)
```

```python
#loop to assign label for pictures

filename=os.listdir(r"C:\aa projects\machine learning\Image data -Deep Learning\dog vs cat\kagglecatsanddogs_3367a\PetImages\train_data_cat_dog")

labels=[]


#cat-->0
#dog--1
for i in range(2000):
    file_name=filename[i]
    label=file_name[0:3]
  
    
    if(label=="cat"):
        labels.append(0)
    else:
        labels.append(1)
        
```

Train test split

```python
#train and test data


X=dog_cat_images
Y=np.asarray(labels)

```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)

```

```python
#Scalig the data
X_train_scaled=X_train/255
X_test_scaled=X_test/255

```

```python
print(X_train_scaled)
print(X_test_scaled)

```

Building the Neural Network


```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
model_mobilenet = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

pretrained_model = hub.KerasLayer(model_mobilenet, input_shape=(224, 224, 3), trainable=False)

```

```python
model=tf.keras.Sequential([
    pretrained_model,
    tf.keras.layers.Dense(2)
])

print(model.summary())

```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
        )

```

```python
model.fit(X_train_scaled,Y_train,epochs=10)
```

```python
score,accuracy = model.evaluate(X_test_scaled,Y_test)
print(score)
print(accuracy)
```
